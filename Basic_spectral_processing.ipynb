### Spectral Processing Example

```python
from sklearn import preprocessing #data processing module
import numpy as np #fast math
from matplotlib import pyplot as plt #plotting utilities 
```

```python
plt.rcParams['figure.figsize'] = [10, 5] #figure size
```

```python
sample = np.array([0, 1.0, 0, 2, 2, 2, 3, 5, 7, 7 , 8, 10, 12, 11, 9, 8, 8, 7, 5, 3, 3, 1, 9, 9, 3, 1, 0, 1]) #array

plt.plot(sample) 
```

```python
sample_scaled = preprocessing.scale(sample) #standardizes, mean = 0, stdev = 1
plt.plot(sample, 'r') #plot original as red
plt.plot(sample_scaled, 'g') #plot standardized as green
```

### Scale a "toy" dataset to the [0, 1] range

```python
sample.shape
```

```python
sample_array = np.tile(sample, [5, 1]) #takes sample and creates an 2d array with 5 of the "sample" array
sample_array.shape
```

```python
sample_array = np.transpose(sample_array, [1, 0]) #takes the transpose
sample_array.shape
```

```python

min_max_scaler = preprocessing.MinMaxScaler() #make a new min_max_scaler object
sample_train_minmax = min_max_scaler.fit_transform(sample_array) #i guess this transforms it to [0,1]?

plt.plot(sample, 'r') 
plt.plot(sample_train_minmax[:, 1], 'g') #processed data
```
